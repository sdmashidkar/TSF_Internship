```python
# Importing libraries in Python
import sklearn.datasets as datasets
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the iris dataset
iris=pd.read_csv('Iris1.csv')
print(iris.head(5))
```

```python
#prints no. of row and columns
print(iris.shape) 
#prints name of columns
print(iris.columns)
feature_names=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
```

```python
import plotly.express as px
fig = px.scatter_3d(iris, x='SepalLengthCm', y='SepalWidthCm', z='PetalWidthCm',color='Species')
fig.show()
```

Here we have used sepal length on the x-axis, sepal width on the y-axis and petal length on the z-axis.

```python
sns.set_style("whitegrid");
sns.pairplot(iris,hue="Species",size=3);
plt.show()
```

```python
sns.FacetGrid(iris,hue="Species",size=5) \
    .map(sns.distplot,"SepalLengthCm") \
    .add_legend()
    
plt.show()
```

```python
X=iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
Y=iris['Species']    
```

### Now let us define the Decision Tree Algorithm

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state =1)
```

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(random_state=1234)
classifier.fit(X_train,y_train)


print('Decision Tree Classifer Created')
```

checking Accuracy

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,y_pred)
print("accuracy score:",acc)
```

### Let us visualize the Decision Tree to understand it better.



```python
from sklearn import tree
fig=plt.figure(figsize=(20,10))
fig=tree.plot_tree(classifier,feature_names=iris.columns,filled='True')
plt.show()
```
